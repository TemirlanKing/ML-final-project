```python
import pandas as pd

positions = {'ST', 'CF', 'LW', 'RW', 'RM', 'LM', 'CAM', 'CDM', 'CM', 'LWB', 'LB', 'RWB', 'RB', 'CB', 'GK'}

def count_positions(alt_pos, positions):
    if pd.notna(alt_pos):  # Check if "Alt Pos" is not NaN
        split_pos = [pos for pos in positions if pos in alt_pos]
        return len(split_pos)
    else:
        return 1


df['Num Positions'] = df['Alt Pos'].apply(count_positions, positions=positions)
df.head(35)
```

```python
df = df.drop(['Alt Pos'], axis=1)
```
